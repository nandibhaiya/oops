```python
#answer 2

from abc import ABC, abstractmethod
import math
class Shape(ABC):
    @abstractmethod
    def area(self):
        pass
class Circle(Shape):
  def __init__(self,radius):
    self.radius = radius
  def area(self):
    return math.pi * self.radius**2
class Rectangle(Shape):
  def __init__(self,length,width):
    self.length = length
    self.width = width
  def area(self):
    return self.length * self.width
circle = Circle(5)
rectangle = Rectangle(4,6)
print(circle.area())
print(rectangle.area())
```

```python
#  answer 3
class Vehicle:
    """
    Base class representing a general vehicle.
    """
    def __init__(self, vehicle_type):
        self.type = vehicle_type
        print(f"Vehicle initialized with type: {self.type}")

    def display_vehicle_info(self):
        return f"This is a general {self.type}."

class Car(Vehicle):
    """
    Derived class from Vehicle, representing a car.
    """
    def __init__(self, vehicle_type, doors):
        # Call the constructor of the base class (Vehicle)
        super().__init__(vehicle_type)
        self.doors = doors
        print(f"Car initialized with {self.doors} doors.")

    def display_car_info(self):
        # Accesses attributes from both Vehicle and Car
        return f"This is a {self.type} with {self.doors} doors."

class ElectricCar(Car):
    """
    Derived class from Car, representing an electric car.
    It adds the 'battery' attribute.
    """
    def __init__(self, vehicle_type, doors, battery_capacity):
        # Call the constructor of the intermediate class (Car)
        super().__init__(vehicle_type, doors)
        self.battery = battery_capacity
        print(f"Electric Car initialized with {self.battery} kWh battery.")

    def display_electric_car_info(self):
        # Accesses attributes from Vehicle, Car, and ElectricCar
        car_info = self.display_car_info()
        return f"{car_info} It is an electric vehicle with a {self.battery} kWh battery."

# --- Usage Example ---

print("--- Creating an ElectricCar object ---")
my_tesla = ElectricCar(vehicle_type="Sedan", doors=4, battery_capacity=75)
print("\n--- Displaying Information ---")
print(my_tesla.display_electric_car_info())
print(f"Type attribute inherited from Vehicle: **{my_tesla.type}**")
print(f"Doors attribute inherited from Car: **{my_tesla.doors}**")
print(f"Battery attribute specific to ElectricCar: **{my_tesla.battery}**")
```

```python
#answer4
class Bird:
    """
    Base class with a common method 'fly()'.
    """
    def fly(self):
        # Default behavior for a bird
        return "This bird is generally flying high in the air."

class Sparrow(Bird):
    """
    Derived class that overrides the fly() method for specific behavior.
    """
    def fly(self):
        # Specialized behavior for a Sparrow
        return "The Sparrow is **flitting** and darting between trees."

class Penguin(Bird):
    """
    Derived class that overrides the fly() method to show a non-flying behavior.
    """
    def fly(self):
        # Specialized behavior for a Penguin (cannot fly)
        return "The Penguin **cannot fly**; it is waddling and swimming gracefully."

# --- Polymorphic Usage Example ---

def make_it_fly(bird_object):
    """
    A function that accepts any object derived from Bird and calls its fly() method.
    The output will vary polymorphically based on the object's actual class.
    """
    print(f"{type(bird_object).__name__}: {bird_object.fly()}")

# Creating instances of the different classes
common_bird = Bird()
my_sparrow = Sparrow()
my_penguin = Penguin()

print("--- Calling the same method 'fly()' on different objects ---")
print(f"Bird: {common_bird.fly()}")
print(f"Sparrow: {my_sparrow.fly()}")
print(f"Penguin: {my_penguin.fly()}")

print("\n--- Demonstrating Polymorphism via a common function ---")
# The function call remains the same, but the behavior is specific to the object.
make_it_fly(my_sparrow)
make_it_fly(my_penguin)
```

```python
#answer 5
class BankAccount:
    """
    A class demonstrating encapsulation by managing a private balance
    and exposing controlled methods for transactions.
    """

    def __init__(self, initial_balance=0):
        # The balance is prefixed with __ (double underscore), making it a
        # "name-mangled" attribute. This signals that it should be treated
        # as private and prevents direct external modification.
        self.__balance = initial_balance
        print(f"Account created. Initial balance: ${self.__balance:.2f}")

    # Public method to deposit money
    def deposit(self, amount):
        """Adds a specified amount to the balance."""
        if amount > 0:
            self.__balance += amount
            print(f"Deposit successful: ${amount:.2f}")
            self.check_balance()
        else:
            print("Error: Deposit amount must be positive.")

    # Public method to withdraw money
    def withdraw(self, amount):
        """Subtracts a specified amount from the balance, if funds are available."""
        if amount <= 0:
            print("Error: Withdrawal amount must be positive.")
        elif amount > self.__balance:
            print(f"Error: Insufficient funds. Current balance: ${self.__balance:.2f}")
        else:
            self.__balance -= amount
            print(f"Withdrawal successful: ${amount:.2f}")
            self.check_balance()

    # Public method to check the balance
    def check_balance(self):
        """Provides controlled access to view the private balance."""
        print(f"Current Balance: ${self.__balance:.2f}")

# --- Demonstration of Encapsulation ---

# 1. Create a bank account instance
account = BankAccount(initial_balance=1000)

print("\n--- Performing Transactions via Public Methods ---")
# 2. Deposit money
account.deposit(500)

# 3. Withdraw money successfully
account.withdraw(200)

print("\n--- Demonstrating Encapsulation and Controlled Access ---")
# 4. Attempt to withdraw too much (controlled access prevents this)
account.withdraw(2000)

# 5. Attempt to deposit a negative amount (controlled access prevents this)
account.deposit(-50)

print("\n--- Direct Access Attempt (Encapsulation Guardrail) ---")
# 6. Attempt to directly access or modify the private attribute
try:
    print(f"Attempting to read balance directly: {account.__balance}")
except AttributeError as e:
    print(f"Error when accessing '__balance' directly: {e}")
    print("This shows the private attribute is protected from direct external access.")

# 7. Use the controlled, public method to check the balance again
account.check_balance()

# Note: Python performs 'name mangling', changing '__balance' to '_BankAccount__balance'.
# While an experienced user could still access it via the mangled name, the double
# underscore clearly communicates that this attribute is intended to be private.
```

```python
#answer 6
class Instrument:
    """
    Base class representing a general musical instrument.
    Defines the common method play().
    """
    def play(self):
        # Default behavior
        print("The instrument makes a sound.")

class Guitar(Instrument):
    """
    Derived class that overrides play() to describe a guitar's sound.
    """
    def play(self):
        # Specialized behavior for Guitar
        print("The Guitar is strummed, producing a melodic chord progression.")

class Piano(Instrument):
    """
    Derived class that overrides play() to describe a piano's sound.
    """
    def play(self):
        # Specialized behavior for Piano
        print("The Piano keys are pressed, playing a rich, sonorous scale.")

# --- Polymorphic Function Demonstration ---

def make_music(instrument):
    """
    This function demonstrates runtime polymorphism. It accepts any object
    that is a type of Instrument and calls its play() method.
    The specific version of play() executed is determined at runtime.
    """
    print(f"\n--- Playing a {type(instrument).__name__} ---")
    instrument.play()

# Creating instances of the derived classes
acoustic_guitar = Guitar()
grand_piano = Piano()
generic_instrument = Instrument()

# Demonstrate runtime polymorphism by calling the same function
# with different object types.
make_music(acoustic_guitar)
make_music(grand_piano)
make_music(generic_instrument)

# Directly calling the methods also shows the specific implementation
print("\n--- Direct Method Calls ---")
acoustic_guitar.play()
grand_piano.play()
```

```python
#answer 7
class MathOperations:
    """
    A class demonstrating the use of class methods and static methods
    for mathematical operations.
    """

    # --- Class Method ---
    @classmethod
    def add_numbers(cls, a, b):
        """
        A class method to add two numbers.
        It takes 'cls' as the first argument, giving it access to the class
        (though not strictly necessary for this simple calculation).
        """
        print(f"--- Calling class method from {cls.__name__} ---")
        result = a + b
        return result

    # --- Static Method ---
    @staticmethod
    def subtract_numbers(a, b):
        """
        A static method to subtract two numbers.
        It does not take 'self' or 'cls', as it operates independently
        of the class instance or the class itself.
        """
        print(f"--- Calling static method ---")
        result = a - b
        return result

# --- Demonstration ---

print("Demonstrating Class Method (add_numbers)")
# Class methods are typically called directly on the Class
sum_result = MathOperations.add_numbers(15, 7)
print(f"15 + 7 = {sum_result}")

print("\nDemonstrating Static Method (subtract_numbers)")
# Static methods are also typically called directly on the Class
diff_result = MathOperations.subtract_numbers(25, 10)
print(f"25 - 10 = {diff_result}")

print("\nDemonstration using an instance (both methods work)")
instance = MathOperations()
instance_sum = instance.add_numbers(3, 4)
instance_diff = instance.subtract_numbers(10, 2)

print(f"3 + 4 (via instance) = {instance_sum}")
print(f"10 - 2 (via instance) = {instance_diff}")
```

```python
#answer 8
class Person:
    """
    A class representing a person, which uses a class variable and a
    class method to track the total number of instances created.
    """
    # 1. Class Variable: Tracks the count across all instances.
    total_persons = 0

    def __init__(self, name):
        self.name = name
        # 2. Increment the class variable every time a new person is created
        Person.total_persons += 1
        print(f"Created new Person: {self.name}")

    # 3. Class Method: Used to access the class variable (total_persons)
    # The 'cls' argument refers to the class itself (Person).
    @classmethod
    def get_total_persons(cls):
        """
        Returns the total number of Person objects created so far.
        """
        print("--- Calling class method to check count ---")
        return cls.total_persons

# --- Demonstration ---

# Initially, the count should be 0 (called directly on the class)
print(f"Initial count: {Person.get_total_persons()}")

# Create the first instance
person1 = Person("Alice")

# Check the count after creating one person
print(f"Current count: {Person.get_total_persons()}")

# Create more instances
person2 = Person("Bob")
person3 = Person("Charlie")

print("\n--- Final Count Check ---")
# The class method can also be called via an instance, but it still refers to the class variable.
final_count = person3.get_total_persons()
print(f"Total number of persons created: {final_count}")
```

```python
# answer 9
class Fraction:
    """
    A class representing a mathematical fraction with custom string output.
    """
    def __init__(self, numerator, denominator):
        # Basic validation to prevent division by zero
        if denominator == 0:
            raise ValueError("Denominator cannot be zero.")

        self.numerator = numerator
        self.denominator = denominator

    # Override the built-in __str__ method
    def __str__(self):
        """
        Returns the user-friendly string representation of the fraction
        in the format "numerator/denominator".
        """
        return f"{self.numerator}/{self.denominator}"

    # It's also good practice to override __repr__ (the developer-friendly representation)
    def __repr__(self):
        """
        Returns the official string representation, typically useful for debugging.
        """
        return f"Fraction({self.numerator}, {self.denominator})"

# --- Demonstration ---

# 1. Create a Fraction object
fraction_a = Fraction(3, 4)
fraction_b = Fraction(10, 5)

print("--- Printing Fractions (calls __str__ automatically) ---")

# When you use print() or str(), the __str__ method is used.
print(f"Fraction A: {fraction_a}")
print(f"Fraction B: {fraction_b}")

print("\n--- Direct Conversion to String ---")
# Explicitly calling str() uses the custom __str__ method.
str_a = str(fraction_a)
print(f"Fraction A as string: {str_a}")

print("\n--- Developer Representation (calls __repr__) ---")
# Calling repr() gives the more technical representation
repr_a = repr(fraction_a)
print(f"Fraction A representation: {repr_a}")
```

```python
# answer 10
class Vector:
    """
    A class representing a 2D vector, demonstrating operator overloading
    for the addition operation (+).
    """
    def __init__(self, x, y):
        self.x = x
        self.y = y

    # Override the built-in __add__ method to enable the '+' operator
    def __add__(self, other):
        """
        Performs vector addition (self + other).
        The components are added element-wise: (x1 + x2, y1 + y2).
        """
        # Ensure the 'other' object is also a Vector
        if not isinstance(other, Vector):
            raise TypeError("Can only add a Vector to another Vector.")

        # Calculate the new components
        new_x = self.x + other.x
        new_y = self.y + other.y

        # Return a new Vector object as the result
        return Vector(new_x, new_y)

    # Override __str__ for clean output
    def __str__(self):
        """
        Returns a user-friendly string representation of the vector.
        """
        return f"Vector({self.x}, {self.y})"

# --- Demonstration ---

# 1. Create two Vector objects
vector_a = Vector(x=3, y=5)
vector_b = Vector(x=-1, y=2)

print("--- Initial Vectors ---")
print(f"Vector A: {vector_a}")
print(f"Vector B: {vector_b}")

# 2. Use the '+' operator (which calls the custom __add__ method)
# This is where operator overloading is demonstrated.
vector_c = vector_a + vector_b

print("\n--- Result of Vector Addition (A + B) ---")
print(f"Vector C: {vector_c}")

# Verification of the result: (3 + -1) = 2, (5 + 2) = 7
print(f"\nVerification: C.x = {vector_c.x}, C.y = {vector_c.y}")

# 3. Demonstration of chaining operations
vector_d = Vector(1, 1)
vector_sum = vector_a + vector_b + vector_d
print(f"\nResult of (A + B + D): {vector_sum}")
```

```python
#ansewer 12
class Student:
    """
    Represents a student with a name and a list of grades,
    and includes a method to calculate the average grade.
    """
    def __init__(self, name, grades):
        """
        Initializes the Student object.
        :param name: The name of the student (string).
        :param grades: A list of numerical grades (list of int/float).
        """
        self.name = name
        self.grades = grades

    def average_grade(self):
        """
        Computes and returns the average of the student's grades.
        Returns 0.0 if the grades list is empty.
        """
        if not self.grades:
            return 0.0

        # Calculate the sum of all grades and divide by the number of grades
        total_sum = sum(self.grades)
        count = len(self.grades)

        return total_sum / count

# --- Demonstration ---

# 1. Create student instances
student_a = Student("Elara Vance", [88, 92, 79, 95, 84])
student_b = Student("Kai Lee", [70, 75, 68, 80])
student_c = Student("Empty Grades", [])

print(f"Student: {student_a.name}")
print(f"Grades: {student_a.grades}")
# Calculate and format the average
avg_a = student_a.average_grade()
print(f"Average Grade: {avg_a:.2f}")

print("\n-------------------------")

print(f"Student: {student_b.name}")
print(f"Grades: {student_b.grades}")
avg_b = student_b.average_grade()
print(f"Average Grade: {avg_b:.2f}")

print("\n-------------------------")
# Test case for an empty list
print(f"Student: {student_c.name}")
print(f"Grades: {student_c.grades}")
avg_c = student_c.average_grade()
print(f"Average Grade: {avg_c:.2f} (Correctly returns 0.0)")
```

```python
# answer 13
class Rectangle:
    """
    A class to represent a rectangle, allowing dimensions to be set
    and the area to be calculated.
    """
    def __init__(self, width=0, height=0):
        """
        Initializes the Rectangle with optional width and height.
        """
        self.width = width
        self.height = height

    def set_dimensions(self, width, height):
        """
        Sets the width and height of the rectangle.
        Ensures dimensions are non-negative.
        """
        if width >= 0 and height >= 0:
            self.width = width
            self.height = height
            print(f"Dimensions set to Width: {self.width}, Height: {self.height}")
        else:
            print("Error: Dimensions must be non-negative.")

    def area(self):
        """
        Calculates and returns the area of the rectangle (Width * Height).
        """
        return self.width * self.height

    def __str__(self):
        """
        Provides a friendly string representation.
        """
        return f"Rectangle(Width={self.width}, Height={self.height})"

# --- Demonstration ---

# 1. Create a Rectangle object
rect1 = Rectangle()
print(f"Initial State: {rect1}")

# 2. Set dimensions using the set_dimensions method
print("\n--- Setting Dimensions for rect1 ---")
rect1.set_dimensions(width=5.5, height=4.0)

# 3. Calculate and display the area
calculated_area = rect1.area()
print(f"Calculated Area of rect1: {calculated_area:.2f}") # Area: 22.00

# 4. Create another rectangle using initial parameters
rect2 = Rectangle(10, 3)
print(f"\nSecond Rectangle: {rect2}")
print(f"Calculated Area of rect2: {rect2.area():.2f}") # Area: 30.00
```

```python
#answer 14
class Employee:
    """
    Base class for all employees.
    Calculates salary based on hourly rate and hours worked.
    """
    def __init__(self, name, hourly_rate, hours_worked):
        self.name = name
        self.hourly_rate = hourly_rate
        self.hours_worked = hours_worked

    def calculate_salary(self):
        """
        Computes the base salary.
        """
        base_salary = self.hourly_rate * self.hours_worked
        return base_salary

    def display_salary_details(self):
        """Helper method to print details for general employees."""
        salary = self.calculate_salary()
        return f"Employee: {self.name}, Base Salary: ${salary:.2f}"

class Manager(Employee):
    """
    Derived class from Employee.
    Adds a fixed bonus to the calculated base salary.
    """
    def __init__(self, name, hourly_rate, hours_worked, bonus):
        # Call the base class constructor to initialize common attributes
        super().__init__(name, hourly_rate, hours_worked)
        self.bonus = bonus

    def calculate_salary(self):
        """
        Overrides the base method. Computes base salary and adds the bonus.
        """
        # Call the base class's calculate_salary method to get the foundation
        base_salary = super().calculate_salary()
        total_salary = base_salary + self.bonus
        return total_salary

    def display_salary_details(self):
        """Helper method to print details for managers."""
        total_salary = self.calculate_salary()
        base_salary = super().calculate_salary()
        return (
            f"Manager: {self.name}, Base Salary: ${base_salary:.2f}, "
            f"Bonus: ${self.bonus:.2f}, Total Salary: ${total_salary:.2f}"
        )

# --- Demonstration ---

print("--- Calculating Employee Salary ---")
employee = Employee("Sarah Connor", hourly_rate=25.0, hours_worked=160)
print(employee.display_salary_details())

print("\n--- Calculating Manager Salary ---")
manager = Manager("John Reese", hourly_rate=35.0, hours_worked=160, bonus=2000.0)
print(manager.display_salary_details())

print("\n--- Verification ---")
# Employee salary: 25.0 * 160 = $4000.00
# Manager salary: (35.0 * 160) + 2000.0 = 5600.00 + 2000.00 = $7600.00
```

```python
#answer 15
class Product:
    """
    Represents a product item with name, unit price, and quantity.
    Includes a method to calculate the total price for the stock held.
    """
    def __init__(self, name, price, quantity):
        """
        Initializes the Product object.
        :param name: Name of the product (string).
        :param price: Unit price of the product (float).
        :param quantity: Stock quantity of the product (integer).
        """
        self.name = name
        self.price = price
        self.quantity = quantity

    def total_price(self):
        """
        Calculates the total price of the product inventory by multiplying
        the unit price by the quantity.
        """
        # Ensure that price and quantity are valid for calculation
        if self.price < 0 or self.quantity < 0:
            print("Warning: Price and quantity must be non-negative.")
            return 0.0

        return self.price * self.quantity

    def __str__(self):
        """
        Provides a friendly string representation.
        """
        return f"Product: {self.name} | Unit Price: ${self.price:.2f} | Quantity: {self.quantity}"

# --- Demonstration ---

# 1. Create product instances
product_a = Product("Laptop Charger", 49.99, 15)
product_b = Product("Wireless Mouse", 12.50, 100)
product_c = Product("Display Cable", 5.00, 0) # Zero stock

print("--- Product Inventory Totals ---")

# Calculate and display total price for Product A
print(product_a)
total_a = product_a.total_price()
print(f"Total Inventory Value: ${total_a:.2f}\n")

# Calculate and display total price for Product B
print(product_b)
total_b = product_b.total_price()
print(f"Total Inventory Value: ${total_b:.2f}\n")

# Calculate and display total price for Product C
print(product_c)
total_c = product_c.total_price()
print(f"Total Inventory Value: ${total_c:.2f}")
```

```python
#answer 16
import abc
from abc import ABC, abstractmethod

class Animal(ABC):
    """
    Abstract Base Class (ABC) for all animals.
    It defines the necessary interface (the sound() method) but does not implement it.
    """
    def __init__(self, name):
        self.name = name

    @abstractmethod
    def sound(self):
        """
        An abstract method that must be implemented by all derived classes.
        """
        # This method has no body in the abstract class.
        pass

class Cow(Animal):
    """
    Concrete class derived from Animal. Implements the sound() method.
    """
    def sound(self):
        """
        The Cow makes a distinct sound.
        """
        return f"{self.name} the Cow says: Moo!"

class Sheep(Animal):
    """
    Concrete class derived from Animal. Implements the sound() method.
    """
    def sound(self):
        """
        The Sheep makes a distinct sound.
        """
        return f"{self.name} the Sheep says: Baa!"

# --- Demonstration of Abstract Class and Polymorphism ---

def make_animal_sound(animal: Animal):
    """
    A function that accepts any object derived from Animal and calls its sound() method.
    This demonstrates polymorphism.
    """
    print(animal.sound())

# 1. Create instances of the concrete (non-abstract) classes
cow = Cow("Bessie")
sheep = Sheep("Shaun")

print("--- Animal Sounds (Polymorphic Behavior) ---")
make_animal_sound(cow)
make_animal_sound(sheep)

print("\n--- Direct Calls ---")
print(cow.sound())
print(sheep.sound())

# 2. Attempting to instantiate the Abstract Base Class directly will result in an error
try:
    abstract_animal = Animal("Unknown")
except TypeError as e:
    print(f"\n--- Abstract Class Restriction ---")
    print(f"Error caught: {e}")
    print("This confirms that you cannot instantiate an abstract class with unimplemented methods.")
```

```python
#answer 17
class Book:
    """
    Represents a book with attributes for title, author, and publication year.
    """
    def __init__(self, title, author, year_published):
        """
        Initializes the Book object.
        :param title: The title of the book (string).
        :param author: The author of the book (string).
        :param year_published: The year the book was published (integer).
        """
        self.title = title
        self.author = author
        self.year_published = year_published

    def get_book_info(self):
        """
        Returns a nicely formatted string containing all the book's details.
        """
        return f'"{self.title}" by {self.author} (Published: {self.year_published})'

# --- Demonstration ---

# 1. Create instances of the Book class
book1 = Book(
    title="The Hitchhiker's Guide to the Galaxy",
    author="Douglas Adams",
    year_published=1979
)

book2 = Book(
    title="Pride and Prejudice",
    author="Jane Austen",
    year_published=1813
)

print("--- Book Information ---")

# 2. Call the get_book_info method to retrieve and print details
info1 = book1.get_book_info()
print(f"Book 1: {info1}")

info2 = book2.get_book_info()
print(f"Book 2: {info2}")
```

```python
#answer 18
class House:
    """
    Base class representing a general residential property.
    """
    def __init__(self, address, price):
        """
        Initializes the House object.
        :param address: The physical location of the property (string).
        :param price: The selling price of the property (float).
        """
        self.address = address
        self.price = price
        print(f"House created at {self.address}.")

    def display_info(self):
        """Returns basic information about the house."""
        return f"Address: {self.address}, Price: ${self.price:,.2f}"

class Mansion(House):
    """
    Derived class from House, adding a specific attribute for number of rooms.
    """
    def __init__(self, address, price, number_of_rooms):
        # Call the base class (House) constructor using super()
        super().__init__(address, price)
        self.number_of_rooms = number_of_rooms
        print(f"Mansion created with {self.number_of_rooms} rooms.")

    def display_info(self):
        """
        Overrides the base method to include the number of rooms.
        """
        base_info = super().display_info()
        return f"{base_info}, Rooms: {self.number_of_rooms}"

# --- Demonstration ---

print("--- Creating a standard House object ---")
house1 = House(address="123 Main St", price=350000.00)
print(house1.display_info())

print("\n--- Creating a Mansion object ---")
# The Mansion instance automatically inherits 'address' and 'price'
mansion1 = Mansion(
    address="789 Summit Drive",
    price=7500000.00,
    number_of_rooms=45
)
print(mansion1.display_info())

# Accessing inherited attributes directly:
print(f"\nPrice of Mansion: ${mansion1.price:,.2f}")
print(f"Address of Mansion: {mansion1.address}")
```
